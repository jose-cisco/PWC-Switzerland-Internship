# Exploratory Data Analysis (EDA)

```python
n_forecast = 30  # number of forecast periods
n_vintage = df.shape[0]  # number of vintages
discount_rate = 0.025  # discount rate (annual)

historical_cf = df.iloc[:, 2:].values  # matrix of cash flows that have already been observed
amount_originated = df.iloc[:, 1].values  # vector of amounts originated per vintage

periods_remaining = n_forecast - np.arange(1, n_vintage + 1)  # number of periods to be forecasted per vintage

paid_percentages = historical_cf / amount_originated.reshape(-1, 1)  # repayment percentage
first_period = np.diag(paid_percentages)  # repayment percentage in the period that the loans were originated per vintage

# Cash flow percentage in the period after the loans were originated per vintage
second_period = np.concatenate([np.diag(paid_percentages[:-1, 1:]), np.array([paid_percentages[-1, -1] * 2])])

p = np.zeros((n_vintage, n_forecast))  # matrix of zeros
p[:, 0] = first_period  # assign repayment percentage of first period
p[:, 1] = second_period  # assign repayment percentage of second period


```

```python
for i in range(n_vintage):
    for j in range(2, n_forecast):
        p[i, j] = max(0, p[i, 1] * np.log(1 + (1 - np.sum(p[i, 0:j]))) * (1 - (j - 1) / n_forecast))

p_forecast = np.zeros((n_vintage, n_forecast - 1))  # matrix of zeros for the forecasted expected repayment percentages

for i in range(n_vintage):
    for j in range(periods_remaining[i]):
        p_forecast[i, j] = p[i, n_forecast - periods_remaining[i] + j]

discount_factors = 1 / (1 + discount_rate) ** (np.arange(1, n_forecast) / 12)  # series of discount factors
pv = (p_forecast.T * discount_factors) * amount_originated.reshape(-1, 1)  # present value of forecasted cash flows
result = np.sum(pv)  # sum of all present values, i.e. the value of the portfolio

print("The fair value estimate for the portfolio is", round(result, 2), "Swiss Francs")
```
